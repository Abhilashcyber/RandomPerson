```
import random
truthCount= 0

for k in range(1,100):
  dieroll = random.randint(1,6)
  Truth_lie_decider = random.randint(1,6) #if number pops 1 then it is a truth because 1/6 probability anything else is 5/6th probability
  if Truth_lie_decider == 1 and dieroll == 4:
    print('It is a four..')
    truthCount = truthCount + 1
  elif Truth_lie_decider == 1 and dieroll != 4:
    print('It is not a four..')
    truthCount = truthCount + 1
  elif Truth_lie_decider != 1 and dieroll == 4:
    print('It is not a four..')
  elif Truth_lie_decider != 1 and dieroll !=4:
    print('It is a four..')
print(truthCount)
```
